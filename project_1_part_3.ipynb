```python
filename="sales_predictions_2023.csv"
filepath="/content.drive/MyDrive/week 3/sales_predictions_2023.csv"
df=pd.read_csv("/content.drive/MyDrive/week 3/sales_predictions_2023.csv")
print(df)
```

```python
df.info()
```

```python
numerical_features=['Item_Identifier','Item_Fat_Content', 'Item_Weight', 'Item_Visibility', 'Item_Type', 'Item_MRP', 'Outlet_Identifier', 'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', 'Item_Outlet_Sales']
for feature in numerical_features:
    sns.histplot(df[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
numerical_features=['Item_Identifier','Item_Fat_Content', 'Item_Weight', 'Item_Visibility', 'Item_Type', 'Item_MRP', 'Outlet_Identifier', 'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', 'Item_Outlet_Sales']
for feature in numerical_features:
    sns.boxplot(x=feature, data=df)
    plt.title(f'Boxplot of {feature}')
    plt.xlabel(feature)
    plt.show()
```

```python
categorical_features=['Item_Identifier','Item_Fat_Content', 'Item_Weight', 'Item_Visibility', 'Item_Type', 'Item_MRP', 'Outlet_Identifier', 'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type', 'Item_Outlet_Sales']
for feature in categorical_features:
    sns.countplot(x=feature, data=df)
    plt.title(f'Countplot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
